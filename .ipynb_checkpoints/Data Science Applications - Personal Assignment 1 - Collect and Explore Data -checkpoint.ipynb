```python
# Hàm khởi tạo các từ khoá tìm kiếm playlist tuỳ theo nhu cầu khảo sát dữ liệu
# Kết quả trả về là danh sách các đường dẫn của trang tìm kiếm playlist theo các keywords

def generate_keywords():
    key_words = []
    hot_keys = ['All music genres', 'Alternative Rock', 'Ambient', 'Classical', 'Country', 'Dance %26 EDM',
               'Dancehall', 'Deep House', 'Disco', 'Drum %26 Bass', 'Dubstep', 'Electronic', 'Folk %26 Singer-Songwriter',
               'Hip-hop %26 Rap', 'Hip-hop %26 Rap', 'House', 'Indie', 'Jazz %26 Blues', 'Latin', 
               'Metal', 'Piano', 'Pop', 'R%26B %26 Soul', 'Reggae', 'Reggaeton', 'Rock', 
                'Soundtrack', 'Techno', 'Trance', 'Trap', 'Triphop', 'World', 
                'Audiobooks','Business', 'Comedy', 'Entertainment', 'Learning', 'News %26 Politics',
                'Religion %26 Spirituality', 'Science', 'Sports', 'Storytelling', 'Technology',
                'Best Of Chill Nation']

    for key in hot_keys:
            key_words.append('https://soundcloud.com/search/sets?q=' + key)
            
    singers = ['sam smith', 'ed sheeran', 'westlife', 'one direction', 'billie eilish', 'nicki minaj', 
               'madonna', 'mariah carey', 'whitney houston', 'celine dion', 'bruno mars', 'beyonce', 
               'jenifer lopez', 'justin bieber', 'adele', 'lady gaga', 'taylor swift', 'rihanana', 
               'demi lovato', 'ariana grande', 'selena gomez', 'dua lipa', 'katy perry']
        
    for singer in singers:
        key_words.append('https://soundcloud.com/search/sets?q=' + singer)
        
    return key_words
```

```python
def scroll_page(seconds, driver):
    lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
    start_time = time.time()
    
    while(True):
        current_time = time.time()
        elapsed_time = current_time - start_time
        lastCount = lenOfPage
        time.sleep(3)
        lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
        if elapsed_time > seconds:
            break
```

```python
# Hàm lấy danh sách các links playlists cần tìm dựa vào danh sách từ khoá truyền vào
def generate_playlists_links(playlists_urls_file, playlists_search_urls, sleep_time = 1, scroll_page_time = 4):
    playlists = []
    playlist_urls_file = open(playlists_urls_file, 'a', encoding='utf-8')

    driver = webdriver.Chrome(ChromeDriverManager().install())
    
    for url in playlists_search_urls:
        driver.get(url)
        driver.maximize_window()
        time.sleep(sleep_time)

        scroll_page(scroll_page_time, driver)
        count = len(driver.find_elements_by_xpath("//div[@class='soundTitle__usernameTitleContainer']"))

        for i in range(1, count+1):
            try:
                track_counter = driver.find_element_by_xpath("/html/body/div[1]/div[2]/div[2]/div[1]/div[1]/div[3]/div[1]/div[1]/div[1]/ul[1]/li["+str(i)+"]/div[1]/div[1]/div[1]/div[2]/div[3]/div[1]/div[1]/a").text
                check_track_array = re.split(" ",track_counter)
                check_tracks = check_track_array[1]
                if (int(check_tracks) > 5):
                    playlists = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[2]/div[1]/div[1]/div[3]/div[1]/div[1]/div[1]/ul/li['+str(i)+']/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[2]/a[1]')
                    playlist_urls_file.write(playlists.get_attribute("href")+"\n") 
            except:
                continue
    driver.close()
    playlist_urls_file.close()
```

```python
def prepare_playlists_urls():
    playlist_search_urls = generate_keywords()
    playlists_urls = generate_playlists_links('playlist_links.txt', playlist_search_urls, 1, 4)
    
    return playlists_urls
```

```python
def collect_playslist(playlists_file):
    driver = webdriver.Chrome(ChromeDriverManager().install())

    plays_file = open('playlist_links.txt', 'r+')
    urls = plays_file.readlines()

    users_file = open('users_links.txt', 'a', encoding='utf-8')
    tracks_file = open("tracks_links.txt", 'a', encoding='utf-8')
    
    # User Links
    users_links = []    
    
    # Playlist information
    users = []
    playlist_names = []
    time_play = []
    like_counts = []
    repost_counts = []
    timepost = []
    
    temp = 0;
    for url in urls:
        driver.get(url)
        driver.maximize_window()
        time.sleep(10)
        count = 0
        i = 0

        if (temp > 1000):
            break;
        temp = temp + 1;
        users.append(driver.find_element_by_xpath('html/body/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div[2]/div[2]/div[1]/div[1]/div[2]/div[1]/a').text)
        users_links.append(driver.find_element_by_xpath('html/body/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div[2]/div[2]/div[1]/div[1]/div[2]/div[1]/a').get_attribute("href") + "\n")
    
        playlist_names.append(driver.find_element_by_xpath('html/body/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div[2]/div[2]/div[1]/div[1]/div[2]/span[1]/span').text)
        try:
            like_counts.append(driver.find_element_by_xpath('html/body/div[1]/div[2]/div[2]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/ul/li[1]/a/span[2]').text)
        except:
            like_counts.append(0)
            
        try:
            repost_counts.append(driver.find_element_by_xpath('html/body/div[1]/div[2]/div[2]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/ul/li[2]/a/span[2]').text)
        except:
            repost_counts.append(0)
                
        timepost.append(driver.find_element_by_xpath('html/body/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div[2]/div[3]/div[1]/time').get_attribute("title"))

        while(count <= 5 and i < 10):
            try:    
                check_views = driver.find_element_by_xpath("/html/body/div[1]/div[2]/div[2]/div[1]/div[3]/div[1]/div[1]/div[2]/div[2]/div[1]/div[3]/div[1]/ul[1]/li["+str(i)+"]/div[1]/div[4]/span[2]")
                track_link = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[2]/div[1]/div[3]/div[1]/div[1]/div[2]/div[2]/div[1]/div[3]/div[1]/ul[1]/li['+str(i)+']/div[1]/div[3]/a[2]').get_attribute("href")
                tracks_file.write(track_link + "\n")
                i = i + 1
                count = count + 1
            except:
                i = i + 1
                continue  

    driver.close()
    
    users_file.writelines(list(set(users_links)))
    users_file.close()
    tracks_file.close()
    
    df = pd.DataFrame(list(zip(users, playlist_names, like_counts, repost_counts, timepost)), 
                      columns= ['users', 'playlist_names', 'like_counts', 'repost_counts','timepost'])
    df.to_csv('play_lists.csv', header = True)
```

```python
def collect_users(users_file):
    users_links_file = open('users_links.txt', 'r', encoding='utf-8')
    urls = users_links_file.readlines()
    users_links_file.close()
    
    usernames = []
    locations = []
    fullnames = []
    followers = []
    followings = []
    tracks = []
    temp = 0
    driver = webdriver.Chrome(ChromeDriverManager().install())

    for url in urls:
        if (temp > 1000):
            break;
        temp = temp + 1;
        driver.get(url)
        driver.maximize_window()
        time.sleep(5)

        usernames.append(driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div[1]/div[1]/div[2]/h3[1]').text)

        try:
            locations.append(driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div[1]/div[1]/div[2]/h4[2]').text)
        except:
            locations.append(0);
            
        try:
            fullnames.append(driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div[1]/div[1]/div[2]/h4[1]').text)
        except:
            fullnames.append(0)
            
        followers.append(driver.find_element_by_xpath('html/body/div[1]/div[2]/div[2]/div[1]/div[4]/div[2]/div[1]/article[1]/table[1]/tbody[1]/tr[1]/td[1]/a[1]/div[1]').text)
        followings.append(driver.find_element_by_xpath('html/body/div[1]/div[2]/div[2]/div[1]/div[4]/div[2]/div[1]/article[1]/table[1]/tbody[1]/tr[1]/td[2]/a[1]/div[1]').text)
        tracks.append(driver.find_element_by_xpath('html/body/div[1]/div[2]/div[2]/div[1]/div[4]/div[2]/div[1]/article[1]/table[1]/tbody[1]/tr[1]/td[3]/a[1]/div[1]').text)
        
    driver.close()
        
    df = pd.DataFrame(list(zip(usernames, fullnames, locations, followers, followings, tracks)), 
                      columns= ['usernames', 'fullnames', 'locations', 'followers', 'followings','tracks'])
    df.to_csv(users_file, encoding ='utf-8')
```

```python
def collect_tracks(tracks_file):
    tracks_links_file = open("tracks_links.txt", "r+") 
    urls = tracks_links_file.readlines()
    tracks_links_file.close()
    
    users = []
    authors = []
    views = []
    likes = []
    reposts = []
    count_cmts = []
    
    playlists = []
    time_posts = []
    temp = 0;
    
    driver = webdriver.Chrome(ChromeDriverManager().install())

    
    for url in urls:
        if (temp > 1000):
            break;
        temp = temp + 1
        driver.get(url)
        driver.maximize_window()
        time.sleep(5)

        try:
            users.append(driver.find_element_by_xpath("html/body/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div[2]/div[2]/div[1]/div[1]/div[2]/div[1]/a[1]").text)
        except:
            users.append(0)
            
        try:
            authors.append(driver.find_element_by_xpath("html/body/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div[2]/div[2]/div[1]/div[1]/div[2]/span[1]/span[1]").text)
        except:
            authors.append(0)
            
        try:
            views.append(driver.find_element_by_xpath("/html/body/div[1]/div[2]/div[2]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/ul[1]/li[1]").text)
        except:
            views.append(0)
          
        try:
            likes.append(driver.find_element_by_xpath("/html/body/div[1]/div[2]/div[2]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/ul[1]/li[2]").text)
        except:
            likes.append(0);
            
        try:
            reposts.append(driver.find_element_by_xpath("/html/body/div[1]/div[2]/div[2]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/ul[1]/li[3]").text) 
        except:
            reposts.append(0);
            
        try:
            count_cmts.append(driver.find_element_by_xpath("/html/body/div[1]/div[2]/div[2]/div[1]/div[3]/div[1]/div[1]/div[2]/div[2]/div[1]/div[3]/div[1]/h3[1]/span[2]").text)
        except:
            count_cmts.append(0);
            
        try:
            playlists.append(driver.find_element_by_xpath("html/body/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div[2]/div[2]/div[2]/a[1]/div[1]/div[2]/span").text)
        except:
            playlists.append(0)
            
        try:
            time_posts.append(driver.find_element_by_xpath("html/body/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div[2]/div[3]/div[1]/time").get_attribute("title"))
        except:
            time_posts.append(0)
            
    driver.close()
            
    df = pd.DataFrame(list(zip(users, authors, views, likes, reposts, count_cmts, playlists, time_posts)), 
                      columns = ['users', 'authors', 'views', 'likes', 'reposts','count_cmts', 'playlists', 'time_posts'])
   
    df.to_csv(tracks_file, encoding ='utf-8')
```
